```
from yahoofinancials import YahooFinancials
from datetime import date, timedelta, datetime
from dateutil.relativedelta import relativedelta
import pandas as pd
import jsonlib
from pyspark.sql.functions import *
```

```
today = date.today()
lastYear = date(today.year - 1, today.month, today.day)
```

```
cryptocurrencies = ['BTC-USD']                                                                                          # Request for crypto currencie Bitcoin
yahoo_financials_cryptocurrencies = YahooFinancials(cryptocurrencies)
historicalPrice = yahoo_financials_cryptocurrencies.get_historical_price_data(str(lastYear), str(today), 'daily')
print(historicalPrice)
```

```
dataframe = pd.json_normalize(historicalPrice)
display(dataframe)
```

```
df = dataframe['BTC-USD.prices'].explode()
print(df)
```

```
jsonList = jsonlib.loads(df.to_json(orient="records"))
print(jsonList)
```

```
flatTable = pd.json_normalize(jsonList)
print(flatTable)
```

```
sparkDF = spark.createDataFrame(flatTable)
display(sparkDF)
```

```
roundTable = sparkDF.select("*", round("close",2).alias ("Closing_Values"))
display(roundTable)
```

```
finalTable = roundTable.select("formatted_date","Closing_Values").sort(asc("formatted_date"))
display(finalTable)
```

```
finalTable.write.format("delta").saveAsTable("default.bitcoint")
```

```
%sql

with   -- Creating a CTE

masterTable as 

(

select
  date_format(BT.formatted_date, 'MM-yyyy') as Date_Month_Year,
  date_format(BT.formatted_date,'MMMM') as Date_Month,
  date_format(BT.formatted_date,'MM') as MonthNumb,
  date_format(BT.formatted_date, 'yyyy') as Date_Year,
  round(avg(BT.Closing_Values),2) as Monthly_avg,
  round(max(BT.Closing_Values),2) as Monthly_max,
  round(min(BT.Closing_Values),2) as Monthly_min
  
from
  default.bitcoint as BT

group by 1, 2, 3, 4

),

finalTable

(

select
  row_number() over(order by Date_Year asc, MonthNumb asc) as ID_Row,
  Date_Month_Year,
  Date_Month,
  MonthNumb,
  Date_Year,
  Monthly_avg,
  Monthly_max,
  Monthly_min

from
  masterTable

order by
  ID_Row

)

select 
  *,
  case
    when ID_Row < 10 then concat('0', ID_Row, '_', '(', Date_Month_Year, ')')
    else concat (ID_Row, '_', '(', Date_Month_Year, ')')
  end as Data_Row

from
  finalTable
```
